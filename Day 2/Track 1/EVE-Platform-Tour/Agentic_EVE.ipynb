# Key Concepts: From LLMs to Agents

Before diving in, here is a short glossary of the terms you will encounter throughout this notebook.

- **Large Language Model (LLM)**: An AI model trained on massive amounts of text that can understand complex questions, reason across different sources of information, and generate coherent, structured text. Examples: [ChatGPT](https://openai.com/chatgpt) by OpenAI, [Claude](https://claude.ai) by Anthropic, [Gemini](https://gemini.google.com) by Google.

- **General-purpose LLM**: An LLM trained on broad, diverse data (web pages, books, code, forums) so that it can handle a wide range of topics, from writing poetry to explaining quantum physics, without being specialised in any single domain. Examples: [GPT-4o](https://openai.com/index/hello-gpt-4o/) by OpenAI, [Claude Sonnet](https://www.anthropic.com/claude/sonnet) by Anthropic, [Llama](https://llama.meta.com/) by Meta.

- **Domain-specific LLM**: An LLM that has been fine-tuned on data from a particular field so that it understands the field's terminology, conventions, and tasks better than a general-purpose model would. In this notebook we use [**EVE (Earth Virtual Expert)**](https://huggingface.co/eve-esa), an open-weight 24B-parameter model created by **ESA Φ-lab**, **Pi School**, and **Mistral AI**. EVE is fine-tuned from Mistral Small using ~33.5 billion tokens that mix general-domain replay data with synthetic Earth Observation and Earth Sciences content. The result is a model that significantly outperforms its base model and other models of similar size on EO-specific benchmarks, including multiple-choice QA, hallucination resistance, and open-ended domain questions, while maintaining (and even slightly improving) general-purpose capabilities such as math, coding, and instruction following. Importantly, EVE also scores high on tool-calling benchmarks, making it especially well-suited for the agentic workflows we build in this notebook. EVE is completely open-source and released under the Apache 2.0 license. Other examples of domain-specific LLMs: [BioMistral](https://huggingface.co/BioMistral/BioMistral-7B) for biomedical text, [StarCoder 2](https://huggingface.co/bigcode/starcoder2-15b) for code generation.

- **Reasoning**: The ability of an LLM to break a complex question into smaller steps, evaluate intermediate results, and decide what to do next. Rather than producing a single answer from memory, a reasoning model thinks step-by-step, for example: *"I need the fire perimeter first, then I can download the right satellite image, and finally compute the burn severity index."* Examples: the [OpenAI o-series](https://openai.com/index/learning-to-reason-with-llms/) models and [Claude with extended thinking](https://docs.anthropic.com/en/docs/build-with-claude/extended-thinking) are explicitly trained to reason before answering.

- **Tools**: Python functions performing one specific job: converting a place name into coordinates, downloading a satellite image, or querying a fire database. Each tool carries a machine-readable description (name, purpose, expected inputs and outputs) so the LLM can decide which tool to call and with what arguments.

- **Agent**: What you get when you connect an LLM to a set of tools and let it use them in a loop. The agent works through a cycle: *reason* about what information is missing → *call* a tool to get it → *observe* the result → *decide* what to do next. This reason-act-observe loop is called **ReAct** (Reasoning + Acting), and it is the core pattern behind most LLM agents today. Examples: [LangGraph](https://langchain-ai.github.io/langgraph/) (used in this notebook), [OpenAI Agents SDK](https://openai.github.io/openai-agents-python/), [Amazon Bedrock Agents](https://aws.amazon.com/bedrock/agents/).

A standalone LLM has no connection to the outside world: it cannot search information, download a satellite image, or look up yesterday's weather. Its knowledge has a cutoff date, so it knows nothing about recent events. An **agent** solves this by letting the LLM call tools at runtime and reason over their results.

The difference from a traditional EO workflow is fundamental. In a scripted pipeline, you hard-code every step: download imagery, run a fixed processing chain, export results. When the question changes, the code changes. With an agent, the *capabilities* are fixed (geocoding, fire detection, spectral analysis, web search) but the *workflow* is determined at runtime based on what the user asks. The same set of tools can answer "show me fires in Greece in August 2023", "show me a S2 image of Attica in 2018", or "what caused the fire near Narbonne?", each triggering a different tool chain with no hard-coded orchestration.

# Part I - From services to tools

In this section, we explore how to take existing geospatial services and make them accessible to a large language model by wrapping them as **tools**. Tools are Python functions that an agent can invoke autonomously whenever it determines they are useful for accomplishing a specific task.

In the Earth Observation domain, a rich ecosystem of service providers exposes geospatial data through a variety of protocols and APIs. Here are some key examples:

- **[Copernicus Data Space Ecosystem (CDSE)](https://dataspace.copernicus.eu/):** A catalogue of satellite imagery and environmental data that can be queried using four different REST-based protocols, including [OGC STAC](https://stacspec.org/) and [OpenSearch](https://documentation.dataspace.copernicus.eu/APIs/OpenSearch.html).
- **[European Forest Fire Information System (EFFIS)](https://effis.jrc.ec.europa.eu/):** Provides maps and verified information about fire events across Europe, accessible via the [WMS protocol](https://www.ogc.org/standard/wms/).
- **[Google Earth Engine (GEE)](https://earthengine.google.com/):** Offers access to a vast planetary-scale geospatial catalogue through both a [REST API](https://developers.google.com/earth-engine/reference/rest) and a [Python client library](https://developers.google.com/earth-engine/guides/python_install).
- **[NASA Earthdata](https://www.earthdata.nasa.gov/):** A portal providing access to NASA's Earth science datasets, queryable via [CMR (Common Metadata Repository)](https://www.earthdata.nasa.gov/learn/find-data/cmr) APIs covering atmosphere, land, ocean, and cryosphere data.
- **[Sentinel Hub](https://www.sentinel-hub.com/):** A cloud-based service for processing and analysing Sentinel satellite data, accessible via a [Python library](https://sentinelhub-py.readthedocs.io/) and a [Process API](https://docs.sentinel-hub.com/api/latest/api/process/).

## What makes a good tool?

The goal of creating a tool is to write a **connector function** that retrieves information from one of these services and delivers it to the agent in a usable form. The tricky part is designing tools that the agent can use *effectively*.

A well-designed tool should:

- **Do one thing:** A tool that fetches fire hotspots is easier to reason about, and less error-prone, than one that fetches fire hotspots *and* generates a report *and* sends an alert.
- **Have a clear, descriptive name and docstring:** The agent decides which tool to call based on their descriptions, so clarity is critical.
- **Accept simple, explicit inputs:** Avoid complex nested structures or ambiguous parameters, prefer plain types like strings, floats, and dates.
- **Return structured, interpretable output.** The agent needs to be able to reason over the result, so favour clean JSON or plain text over raw binary payloads.

These principles follow the composability phylosophy: small, focused tools are far more powerful when combined than a single monolithic function that tries to do everything.

## Converting a service into tools

In the following sections, we will walk through a concrete example: wrapping a simple yet useful Earth Observation service into a set of tools that an agent can query. This will illustrate the full pipeline, from raw API to a tool the agent can reason about and invoke autonomously.

## Nominatim

[Nominatim](https://nominatim.openstreetmap.org/) is a free geocoding service built on [OpenStreetMap](https://www.openstreetmap.org/) data. It converts human-readable place names, such as "Greece", "Evia Island", or "Tenerife" into geographic coordinates and bounding boxes, with no registration or API key required.

This tool serves as the **entry point for many spatial queries**. When a user asks about active fires or NDVI indices for a specific location, the model does not inherently know the precise bounding box of that place. Nominatim bridges that gap, turning a natural language place name into the structured coordinates that downstream services like EFFIS and CDSE expect.

We implement two complementary operations:
- **Forward geocoding**: place name → bounding box
- **Reverse geocoding**: lat, long → place name


````python
import json
import logging
import sys
from typing import Any

import httpx
from geopy.geocoders import Nominatim
from mcp.server.fastmcp import FastMCP

logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s [%(levelname)s] %(name)s — %(message)s",
    stream=sys.stderr,
)
logger = logging.getLogger("geocode-mcp")

NOMINATIM_SEARCH_URL = "https://nominatim.openstreetmap.org/search"
NOMINATIM_USER_AGENT = "GeocodeMCPServer"
TIMEOUT = 60

async def _http_get(
    url: str,
    params: dict[str, Any] | None = None,
    headers: dict[str, str] | None = None,
) -> httpx.Response:
    """Perform an async HTTP GET with sensible defaults."""
    async with httpx.AsyncClient(timeout=100, follow_redirects=True) as client:
        resp = await client.get(url, params=params, headers=headers)
        resp.raise_for_status()
        return resp
geolocator = Nominatim(user_agent=NOMINATIM_USER_AGENT, timeout=TIMEOUT)
````

### Tool 1 - Geocode place
This tool allow to convert a place into a geographical bounding box. The tool takes in input:
- **place_name**: the string to be converted into bbox
- **buffer_km**: how many kilometers we want to expand the original bbox
- **limit**: how many candidates results to fetch, a single place can have different name's resolutions (e.g. Rhode's island)

### How it works

The `geocode_place_raw` function below handles forward geocoding:

1. **Query**: Sends the place name to the [Nominatim Search API](https://nominatim.org/release-docs/develop/api/Search/), retrieving up to `limit` candidate results in JSON format.
2. **Parse & reorder**: Nominatim returns bounding boxes as `[south, north, west, east]`. These are reordered to `west, south, east, north` to match the convention expected by EFFIS and CDSE.
3. **Buffer (optional)**: If `buffer_km` is provided, the bounding box is expanded by that distance in all directions. The conversion uses the standard approximation of [~111 km per degree of latitude](https://en.wikipedia.org/wiki/Latitude#Length_of_a_degree_of_latitude).

The function returns a JSON string containing the top result, all candidate results, and their associated metadata (display name, OSM type, class, centroid coordinates).

> **Note on the `~111 km/degree` approximation:** this holds well for latitude, but longitude degrees shrink towards the poles (by a factor of `cos(latitude)`). For high-latitude regions, consider applying `buffer_km / (111.0 * cos(radians(lat)))` for the east–west expansion instead.

````python
async def geocode_place_raw(
    place_name: str,
    buffer_km: float = 0.0,
    limit: int = 5,
) -> str:
    """
    Convert a place name to a bounding box.

    Geocodes a place name (city, region, country, landmark, etc.) into a
    bounding box string in "west,south,east,north" format — the same format
    accepted by get_effis_burnt_areas and other tools.

    Uses the OpenStreetMap Nominatim API (free, no API key required).

    Args:
        place_name: Name of the place to geocode (e.g. "Greece",
                    "Athens", "Evia island", "Peloponnese").
        buffer_km:  Optional buffer in km to expand the bbox (default 0).
        limit:      Maximum number of candidate results to return (default 5).

    Returns:
        JSON with the top result's bbox string (ready to pass to other tools)
        and all candidate matches.
    """
    logger.info("Geocoding place: %s", place_name)

    try:
        resp = await _http_get(
            NOMINATIM_SEARCH_URL,
            params={
                "q": place_name,
                "format": "json",
                "limit": str(limit),
                "addressdetails": "1",
            },
            headers={"User-Agent": NOMINATIM_USER_AGENT},
        )
        results = resp.json()
    except Exception as exc:
        return json.dumps({"error": f"Geocoding failed: {exc}"})

    if not results:
        return json.dumps(
            {
                "error": f"No results found for '{place_name}'.",
                "place_name": place_name,
            }
        )

    buf = buffer_km / 111.0

    def _parse_result(r: dict) -> dict:
        # Nominatim boundingbox is [south, north, west, east]
        bb = r.get("boundingbox", [])
        south, north, west, east = (
            float(bb[0]) - buf,
            float(bb[1]) + buf,
            float(bb[2]) - buf,
            float(bb[3]) + buf,
        )
        return {
            "display_name": r.get("display_name", ""),
            "bbox": f"{west:.4f},{south:.4f},{east:.4f},{north:.4f}",
            "bbox_array": [
                round(west, 4),
                round(south, 4),
                round(east, 4),
                round(north, 4),
            ],
            "lat": float(r.get("lat", 0)),
            "lon": float(r.get("lon", 0)),
            "osm_type": r.get("osm_type", ""),
            "class": r.get("class", ""),
            "type": r.get("type", ""),
        }

    parsed = [_parse_result(r) for r in results]

    return json.dumps(
        {
            "place_name": place_name,
            "top_result": parsed[0],
            "all_results": parsed,
            "total_results": len(parsed),
        }
    )
````

### Tool 2 - Reverse geocode
This tool performs the inverse operation of Tool 1: it converts a pair of geographic coordinates into a human-readable place name. The tool takes in input:
- **latitude**: the latitude of the point to look up, in decimal degrees (WGS84, range `-90` to `90`)
- **longitude**: the longitude of the point to look up, in decimal degrees (WGS84, range `-180` to `180`)

### How it works

The `reverse_geocode_place_raw` function below handles reverse geocoding:

1. **Query**: Sends the `(latitude, longitude)` pair to the [Nominatim Reverse API](https://nominatim.org/release-docs/develop/api/Reverse/) through geopy's [`Nominatim.reverse`](https://geopy.readthedocs.io/en/stable/#geopy.geocoders.Nominatim.reverse) wrapper, which returns the *closest* matching feature on OpenStreetMap.
2. **Extract address**: Reads the formatted human-readable address from the resulting `geopy.location.Location` object via `location.address` (a single comma-separated string from country down to street/house level when available).
3. **Error handling**: Any network or service exception is caught and returned as a JSON `{"error": ...}` payload, so the agent can still reason over the failure instead of crashing the workflow.

The function returns a JSON string containing the original `latitude`, `longitude`, and the resolved `address`.

> **Note on coverage and granularity:** Nominatim only resolves coordinates that fall inside [OpenStreetMap](https://www.openstreetmap.org/) data. Points in the open ocean or in unmapped regions cause `Nominatim.reverse` to return `None`, in which case `location.address` would raise `AttributeError`, consider guarding the access with `if location is None` before serialising. The level of detail (country, city, street, building) is controlled by Nominatim's `zoom` parameter, which defaults to `18` (building level); pass a smaller `zoom` if a coarser address is preferable for the agent's reasoning.


````python
async def reverse_geocode_place_raw(
    latitude: float,
    longitude: float,
) -> str:
    """
    Convert latitude and longitude to a place name (reverse geocoding).

    Uses the OpenStreetMap Nominatim API via geopy.

    Args:
        latitude:  Latitude coordinate.
        longitude: Longitude coordinate.

    Returns:
        JSON with the place name and address details.
    """
    logger.info("Reverse geocoding: %s, %s", latitude, longitude)

    try:
        location = geolocator.reverse(f"{latitude}, {longitude}")
    except Exception as exc:
        return json.dumps({"error": f"Reverse geocoding failed: {exc}"})

    return json.dumps(
        {
            "latitude": latitude,
            "longitude": longitude,
            "address": location.address,
        }
    )
````

## Model-context-protocol (MCP)

Now that we have defined our tools we need a way to connect them to the model and this is where MCP comes in play!

### What is MCP?
**MCP (Model Context Protocol)** is an open protocol that standardizes how AI applications connect to external data sources and tools. Think of it as **"USB-C for AI"**: a universal interface that allows AI models to interact with the world beyond their training data.

Traditional AI applications require custom integrations for every external service they need to access. MCP eliminates this complexity by providing a standardized protocol that enables AI models to dynamically discover and use:

- **Prompts**: pre-defined templates or instructions that guide language model interactions
- **Resources**: structured data or content that provides additional context to the model
- **Tools**: executable functions that allow models to perform actions or retrieve information

Unlike a typical LLM pipeline with statically defined components, MCP enables dynamic discovery of the capabilities offered by the connected server. For example, you can extend functionality by simply adding a new tool to the existing server; no changes to the client are required. This makes the system modular, allowing model capabilities to be extended or updated independently.

### MCP Architecture

MCP organizes applications into three components:

1. **Servers**: Python functions exposed as MCP-compliant tools, resources, and prompts that AI models can invoke
2. **Clients**: Establish connections to MCP servers, handling protocol negotiation and lifecycle management automatically
3. **Apps**: Provide interactive user interfaces that render directly within conversations

### Wrapping Our Functions

Now we transform the raw functions we built above into **MCP tools**. Adding the `@mcp.tool()` decorator makes [FastMCP](https://gofastmcp.com) automatically:

1. **Generate JSON schemas** from function signatures and type hints
2. **Create tool documentation** from docstrings
3. **Validate inputs** before calling the function
4. **Handle errors** and return structured responses
5. **Expose the tool** through the MCP protocol

Each wrapper is a thin shell that delegates to the corresponding `*_raw` function, keeping tool logic separate from protocol plumbing

### FastMCP

[FastMCP](https://gofastmcp.com) is a Python framework that automates the complexity of building MCP servers.

````python
from fastmcp import FastMCP

mcp = FastMCP("Geocode")


@mcp.tool()
async def geocode_place(place_name: str, buffer_km: float = 0.0, limit: int = 5) -> str:
    """Convert a place name to a bounding box.

    Args:
        place_name: Name of the place (e.g. "Greece", "Athens", "Evia island").
        buffer_km: Optional buffer in km to expand the bbox.
        limit: Maximum number of candidate results.
    """
    return await geocode_place_raw(place_name, buffer_km, limit)


@mcp.tool()
async def reverse_geocode_place(
    latitude: float,
    longitude: float,
) -> str:
    """Convert latitude and longitude to a place name (reverse geocoding).

    Args:
        latitude: Latitude coordinate.
        longitude: Longitude coordinate.

    Returns:
        JSON with the place name and address details.
    """
    return await reverse_geocode_place_raw(latitude, longitude)
````

### Run the MCP Server

MCP servers support two transport mechanisms:

1. **STDIO**: for local communication where the host launches the server as a subprocess (used by Claude Desktop, Cursor, etc.)
2. **HTTPS**: for network access where a single server serves multiple clients

In this notebook, we run the server over HTTPS in a daemon thread so it stays alive alongside notebook cells.

````python
import threading
import asyncio
import socket
import time

MCP_HOST = "0.0.0.0"
MCP_PORT = 8005


def _port_is_in_use(port: int, host: str = "127.0.0.1") -> bool:
    """Return True if something is already listening on (host, port)."""
    with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
        s.settimeout(0.5)
        try:
            s.connect((host, port))
            return True
        except (ConnectionRefusedError, socket.timeout, OSError):
            return False


def run_mcp_server_in_thread():
    logging.getLogger("uvicorn.access").setLevel(logging.WARNING)
    asyncio.run(mcp.run(transport="http", host=MCP_HOST, port=MCP_PORT, log_level="warning"))


existing_thread = globals().get("server_thread")
if existing_thread is not None and existing_thread.is_alive():
    print("MCP server already running in this kernel (reusing existing thread)")
elif _port_is_in_use(MCP_PORT):
    raise RuntimeError(
        f"Port {MCP_PORT} is already in use by another process. "
        f"Either stop it (macOS/Linux: `lsof -i :{MCP_PORT}` then `kill <pid>`) "
        f"or change MCP_PORT above."
    )
else:
    server_thread = threading.Thread(target=run_mcp_server_in_thread, daemon=True)
    server_thread.start()
    time.sleep(2)

print(f"MCP server running at http://localhost:{MCP_PORT}/mcp")
print(f"Thread alive: {server_thread.is_alive()}")
````

````python
def get_mcp_servers(
    base_url: str,
    token: str,
    page: int = 1,
    limit: int = 10,
    timeout: int = 10,
):
    url = f"{base_url}/mcp-servers"
    
    headers = {
        "accept": "application/json",
        "Authorization": f"Bearer {token}",
    }

    params = {
        "page": page,
        "limit": limit,
    }

    response = requests.get(url, headers=headers, params=params, timeout=timeout)
    response.raise_for_status()

    resp = response.json()["data"]

    return [
        {
            "server_id": row["id"],
            "server_name": row["name"],
            "description": row["description"],
            "url": row["config"].get("url"),
            "transport": row["config"].get("transport"),
        }
        for row in resp
    ]
````

# Part III - EVE's agentic framework

In this section we will explore the agentic capabilities of EVE. The system connects' the model to the tools presented in the toolkit creating an agent capable of planning and executing tools. To achieve this  We use [LangChain](https://python.langchain.com/docs/introduction/) as the foundation and [LangGraph](https://langchain-ai.github.io/langgraph/): a LangChain extension for building stateful, graph-based agentic workflows, to implement the reasoning loop.

At its core, an agent is a LLM that has been given the ability to take actions (specifically, to call external tools) and then incorporate the results of those actions into its next reasoning step. This creates a loop rather than a single forward pass:

<img src="https://eve-nasa-earthrise-book-assets.s3.us-east-1.amazonaws.com/diagrams/react_loop_flow.svg" alt="The ReAct loop: the LLM reasons, calls a tool, observes the result, and repeats until a final response is ready" style="max-width: 700px;" />

## The ReAct Loop

The specific agent architecture used here is called [ReAct (Reason + Act) (Yao et al., 2023)](https://arxiv.org/abs/2210.03629). The model alternates between two modes:

- **Reasoning**: the model thinks through what it knows and what it needs
- **Acting**: the model emits a structured tool call that the host application executes

LangGraph implements this as a graph with exactly **two nodes**: an LLM node and a tools node - connected in a loop that terminates when the LLM produces a response with no tool calls.

#### Calling the agentic framework
Let's implement now both agentic and non-agentic calls

````python
from requests.exceptions import ChunkedEncodingError, ConnectionError as RequestsConnectionError
from urllib3.exceptions import ProtocolError as Urllib3ProtocolError
import requests


GREEN = "\033[92m"
RED = "\033[91m"
YELLOW = "\033[93m"
CYAN = "\033[96m"
BOLD = "\033[1m"
RESET = "\033[0m"

def generate_agentic(
    payload, base_url: str, token: str, conv_id: str
) -> dict | None:
    """Same GenerationRequest as streaming; uses POST /generate-agentic (JSON body, no SSE)."""
    resp = requests.post(
        f"{base_url}/conversations/{conv_id}/generate-agentic",
        json=payload,
        headers={"Authorization": f"Bearer {token}"},
        timeout=600,
    )
    if resp.status_code != 200:
        print(f"Blocking agentic failed ({resp.status_code}): {resp.text[:800]}")
        return None
    data = resp.json()
    return data


def stream_generate_agentic(
    payload,
    base_url: str,
    token: str,
    conv_id: str,
    max_stream_retries: int = 4,
):

    stream_timeout = (30, 600)  # (connect, read between chunks) seconds

    for attempt in range(max_stream_retries):
        t0 = time.perf_counter()
        token_count = 0
        tool_calls = []
        final_payload = None
        answer_chars = 0
        first_token_latency = None

        try:
            with requests.post(
                f"{base_url}/conversations/{conv_id}/stream-generate-agentic",
                json=payload,
                headers={
                    "Authorization": f"Bearer {token}",
                    "Accept": "text/event-stream",
                    "Connection": "keep-alive",
                },
                stream=True,
                timeout=stream_timeout,
            ) as resp:
                if resp.status_code != 200:
                    print(f"Stream endpoint failed ({resp.status_code}): {resp.text[:500]}")
                    return None

                print()
                for raw_line in resp.iter_lines(chunk_size=16384):
                    if not raw_line:
                        continue
                    line = (
                        raw_line.decode("utf-8")
                        if isinstance(raw_line, bytes)
                        else raw_line
                    )
                    if not line.startswith("data: "):
                        continue
                    payload_str = line[len("data: ") :]

                    # Parse JSON events
                    try:
                        event = json.loads(payload_str)
                    except json.JSONDecodeError:
                        # Plain-text token (non-JSON stream mode)
                        token_count += 1
                        if first_token_latency is None:
                            first_token_latency = time.perf_counter() - t0
                        print(payload_str, end="", flush=True)
                        answer_chars += len(payload_str)
                        continue

                    etype = event.get("type")

                    if etype == "tool_call":
                        tool_calls.append(event.get("content", ""))
                        print(
                            f"\n{YELLOW}  🔍 {event.get('content', '')}{RESET}", flush=True
                        )

                    elif etype == "tool_result":
                        preview = event.get("content", "")[:80]
                        print(f"\n{CYAN}  ↩  result preview: {preview}…{RESET}", flush=True)

                    elif etype == "token":
                        content = event.get("content", "")
                        if first_token_latency is None:
                            first_token_latency = time.perf_counter() - t0
                        token_count += 1
                        answer_chars += len(content)
                        print(content, end="", flush=True)

                    elif etype == "final":
                        final_payload = event
                        print()  # newline after streamed tokens

                    elif etype == "stopped":
                        print("Generation was stopped by the server")
                        break

                    elif etype == "error":
                        print(f"Stream error: {event.get('message')}")

        except requests.exceptions.Timeout:
            print("Request timed out (connect or idle-between-chunks limit).")
            if attempt < max_stream_retries - 1:
                wait = 2**attempt
                print(f"{YELLOW}Retrying in {wait}s ({attempt + 2}/{max_stream_retries})…{RESET}")
                time.sleep(wait)
                continue
            print(f"\n{RED}Stream timed out after {max_stream_retries} attempts.{RESET}")
            return None

        except (ChunkedEncodingError, Urllib3ProtocolError, RequestsConnectionError) as e:
            if attempt < max_stream_retries - 1:
                wait = 2**attempt
                print(
                    f"\n{YELLOW}Stream dropped ({type(e).__name__}: {e}). "
                    f"Retrying in {wait}s ({attempt + 2}/{max_stream_retries})…{RESET}",
                    flush=True,
                )
                time.sleep(wait)
                continue
            print(f"\n{RED}Stream failed after {max_stream_retries} attempts: {e}{RESET}")
            return None

        elapsed = time.perf_counter() - t0
        print()
        print(f"Stream completed in {elapsed:.2f}s")
        print(
            f"First-token latency  : {first_token_latency:.2f}s"
            if first_token_latency
            else "First-token latency: n/a"
        )
        print(f"Tokens/events yielded: {token_count}")
        print(f"Answer chars         : {answer_chars}")
        print(f"Tool calls made      : {len(tool_calls)}")
        for i, tc in enumerate(tool_calls):
            print(f"  [{i + 1}] {tc}")
        if final_payload:
            lats = final_payload.get("latencies", {})
            print(f"generation_latency   : {lats.get('generation_latency', 'n/a')}")
            print(f"total_latency        : {lats.get('total_latency', 'n/a')}")

        if answer_chars == 0:
            print("No answer tokens were streamed.")
        return

    return None
````

## Tracing

Tracing is fundamental to understanding what is happening at runtime. In an agentic framework, the agent can invoke multiple tools and go through several rounds of reasoning before deciding it has gathered enough information to answer the original query. The `agentic` endpoint returns the full execution trace, allowing us to inspect and understand the agent's flow.

Within a trace, two types of steps can be identified:

- **Agent step:** represents the execution of the EVE model. This is the entrypoint of the execution, the query is passed to the model, which then decides what action to take next.
- **Tool step:** represents the execution of a single tool. It receives an input from the model and returns an output back to it.

For each step, the trace also exposes the latency and the data produced at that stage.

The trace is returned as part of the request's response. Let's inspect it:

````python
import html
import json
from typing import Any, Dict, List, Tuple

from IPython.display import HTML, display


def _escape(s: Any) -> str:
    if s is None:
        return ""
    return html.escape(str(s), quote=True)


def _pair_tool_inputs(trace: List[Dict[str, Any]]) -> Dict[int, Any]:
    """Map each tools-step index to the args dict from the preceding agent tool_calls queue."""
    queue: List[Tuple[Any, Any]] = []
    out: Dict[int, Any] = {}
    for i, step in enumerate(trace):
        node = (step.get("node") or "").lower()
        if node == "agent":
            for tc in step.get("tool_calls") or []:
                queue.append((tc.get("name"), tc.get("args")))
        elif node == "tools":
            name = step.get("name")
            args = None
            for j, (qn, qa) in enumerate(queue):
                if qn == name:
                    args = qa
                    queue.pop(j)
                    break
            if args is None and queue:
                args = queue.pop(0)[1]
            out[i] = args
    return out


def visualize_agent_trace(result: dict) -> None:
    """Render a timeline of EVE `trace` steps in the notebook."""
    trace = result.get("trace") or []
    query = result.get("query", "")
    meta = result.get("metadata") or {}
    latencies = meta.get("latencies") or {}
    total_lat = latencies.get("total_latency")

    if not trace:
        display(HTML("<p><em>No trace in this result.</em></p>"))
        return

    step_latencies = [float(s.get("latency_s") or 0) for s in trace]
    total_wall = float(total_lat) if total_lat is not None else sum(step_latencies) or 1e-9

    tool_inputs = _pair_tool_inputs(trace)

    cards: list[str] = []
    cumulative = 0.0
    palette = {
        "agent": ("#7c3aed", "#a78bfa", "EVE agent", "Plans the next action or writes the reply."),
        "tools": ("#0891b2", "#22d3ee", "Tool execution", "Runs one tool and returns its payload."),
    }

    for i, step in enumerate(trace):
        node = (step.get("node") or "step").lower()
        hue, accent, title, subtitle = palette.get(
            node, ("#64748b", "#94a3b8", _escape(node), "")
        )
        lat = float(step.get("latency_s") or 0)
        cumulative += lat
        role = _escape(step.get("role", ""))

        body_parts: list[str] = []

        if node == "agent":
            tcalls = step.get("tool_calls") or []
            if tcalls:
                names = ", ".join(_escape(tc.get("name", "?")) for tc in tcalls)
                body_parts.append(
                    f'<p class="muted">Outgoing tool request(s): <code class="inline">{names}</code></p>'
                    '<p class="hint">Arguments and results are shown under the following <strong>tools</strong> step.</p>'
                )
            content = (step.get("content") or "").strip()
            if content:
                body_parts.append(
                    f'<details class="fold"><summary>Assistant message ({len(content)} characters)</summary>'
                    f'<pre class="msg">{_escape(content)}</pre></details>'
                )
            if not body_parts:
                body_parts.append(
                    '<p class="muted">Deciding next action…</p>'
                )

        elif node == "tools":
            tname = _escape(step.get("name", "tool"))
            body_parts.append(f'<p class="tool-title">Tool <code class="tool-tag">{tname}</code></p>')
            inp = tool_inputs.get(i)
            if inp is not None:
                try:
                    inp_txt = json.dumps(inp, indent=2, ensure_ascii=False)
                except (TypeError, ValueError):
                    inp_txt = str(inp)
                body_parts.append(
                    '<span class="lbl">Input (arguments passed to the tool)</span>'
                    f'<pre class="io-block io-in">{_escape(inp_txt)}</pre>'
                )
            else:
                body_parts.append(
                    '<span class="lbl">Input</span>'
                    '<pre class="io-block io-in muted">(Could not match to a preceding tool call — '
                    "showing output only.)</pre>"
                )
            raw = step.get("content")
            out_txt = raw if isinstance(raw, str) else json.dumps(raw, indent=2, ensure_ascii=False, default=str)
            body_parts.append(
                '<span class="lbl">Output (exact value returned to the model)</span>'
                f'<pre class="io-block io-out">{_escape(out_txt)}</pre>'
            )
        else:
            body_parts.append(
                f'<pre class="msg">{_escape(json.dumps(step, indent=2, default=str))}</pre>'
            )

        cards.append(
            f"""
            <article class="card" style="--accent:{accent};--hue:{hue}">
              <div class="rail"></div>
              <div class="card-inner">
                <header>
                  <span class="step-idx">Step {i + 1}</span>
                  <span class="node-pill" style="border-color:{accent};color:{accent}">{_escape(node)}</span>
                  <span class="role">{role}</span>
                  <span class="lat">{lat:.2f}s</span>
                </header>
                <h3>{title}</h3>
                <p class="sub">{subtitle}</p>
                <div class="body">{"".join(body_parts)}</div>
                <footer>≈ {_escape(f"{cumulative:.2f}")}s elapsed in trace</footer>
              </div>
            </article>
            """
        )

    gantt_rows = []
    t0 = 0.0
    for i, step in enumerate(trace):
        lat = float(step.get("latency_s") or 0)
        left = 100 * t0 / total_wall
        width = max(0.5, 100 * lat / total_wall)
        node = (step.get("node") or "?").lower()
        _, accent, _, _ = palette.get(node, ("#64748b", "#94a3b8", "", ""))
        raw_label = step.get("name") or (step.get("tool_calls") or [{}])[0].get("name") or node
        tip = _escape(f"{raw_label} · {lat:.1f}s")
        gantt_rows.append(
            f'<div class="g-seg" style="left:{left:.1f}%;width:{width:.1f}%;background:{accent}" title="{tip}"></div>'
        )
        t0 += lat

    html_doc = f"""
    <style>
      .eve-trace {{
        font-family: ui-sans-serif, system-ui, -apple-system, "Segoe UI", Roboto, Helvetica, Arial;
        background: linear-gradient(145deg, #0f172a 0%, #1e1b4b 55%, #0f172a 100%);
        color: #e2e8f0;
        border-radius: 16px;
        padding: 1.25rem 1.5rem 1.5rem;
        margin: 0.5rem 0 1rem;
        border: 1px solid rgba(148, 163, 184, 0.25);
        box-shadow: 0 20px 50px rgba(0,0,0,0.35);
      }}
      .eve-trace h2 {{
        margin: 0 0 0.35rem;
        font-size: 1.15rem;
        font-weight: 650;
        letter-spacing: -0.02em;
        background: linear-gradient(90deg, #c4b5fd, #67e8f9);
        -webkit-background-clip: text;
        background-clip: text;
        color: transparent;
      }}
      .eve-trace .q {{
        margin: 0 0 1rem;
        font-size: 0.9rem;
        color: #94a3b8;
      }}
      .eve-trace .q strong {{ color: #e2e8f0; }}
      .gantt {{
        position: relative;
        height: 28px;
        background: rgba(15, 23, 42, 0.8);
        border-radius: 8px;
        margin-bottom: 1.25rem;
        border: 1px solid rgba(100, 116, 139, 0.35);
        overflow: hidden;
      }}
      .gantt .g-seg {{
        position: absolute;
        top: 4px;
        bottom: 4px;
        border-radius: 6px;
        opacity: 0.92;
        box-shadow: 0 0 12px rgba(34, 211, 238, 0.25);
      }}
      .gantt-cap {{
        display: flex;
        justify-content: space-between;
        font-size: 0.75rem;
        color: #64748b;
        margin-bottom: 0.35rem;
      }}
      .timeline {{
        display: flex;
        flex-direction: column;
        gap: 0.85rem;
      }}
      .card {{
        display: flex;
        gap: 0;
        background: rgba(15, 23, 42, 0.65);
        border-radius: 12px;
        overflow: hidden;
        border: 1px solid rgba(148, 163, 184, 0.2);
      }}
      .rail {{
        width: 5px;
        background: linear-gradient(180deg, var(--hue), var(--accent));
        flex-shrink: 0;
      }}
      .card-inner {{ flex: 1; padding: 0.75rem 1rem 0.85rem; min-width: 0; }}
      .card header {{
        display: flex;
        flex-wrap: wrap;
        align-items: center;
        gap: 0.5rem;
        margin-bottom: 0.25rem;
      }}
      .step-idx {{
        font-size: 0.7rem;
        text-transform: uppercase;
        letter-spacing: 0.08em;
        color: #64748b;
        font-weight: 600;
      }}
      .node-pill {{
        font-size: 0.72rem;
        font-weight: 600;
        padding: 0.12rem 0.5rem;
        border-radius: 999px;
        border: 1px solid;
        text-transform: uppercase;
        letter-spacing: 0.04em;
      }}
      .role {{ font-size: 0.78rem; color: #94a3b8; }}
      .lat {{
        margin-left: auto;
        font-variant-numeric: tabular-nums;
        font-weight: 700;
        color: #fbbf24;
        font-size: 0.85rem;
      }}
      .card h3 {{
        margin: 0;
        font-size: 1rem;
        font-weight: 650;
        color: #f1f5f9;
      }}
      .sub {{
        margin: 0.15rem 0 0.5rem;
        font-size: 0.8rem;
        color: #94a3b8;
      }}
      .body {{ font-size: 0.85rem; }}
      .muted {{ color: #64748b; margin: 0.25rem 0; }}
      .hint {{ font-size: 0.78rem; color: #64748b; margin: 0.2rem 0 0.5rem; }}
      code.inline {{ color: #a5f3fc; font-size: 0.85rem; }}
      .lbl {{
        font-size: 0.68rem;
        text-transform: uppercase;
        letter-spacing: 0.06em;
        color: #94a3b8;
        display: block;
        margin: 0.65rem 0 0.3rem;
      }}
      .lbl:first-of-type {{ margin-top: 0.15rem; }}
      .io-block {{
        margin: 0;
        padding: 0.55rem 0.65rem;
        border-radius: 8px;
        background: #020617;
        color: #cbd5e1;
        font-size: 0.72rem;
        line-height: 1.4;
        overflow-x: auto;
        white-space: pre-wrap;
        word-break: break-word;
        border: 1px solid rgba(51, 65, 85, 0.6);
      }}
      .io-in {{ border-left: 3px solid #a78bfa; }}
      .io-out {{ border-left: 3px solid #22d3ee; }}
      .msg {{
        margin: 0.35rem 0 0;
        padding: 0.5rem 0.55rem;
        border-radius: 6px;
        background: #020617;
        color: #cbd5e1;
        font-size: 0.78rem;
        line-height: 1.45;
        overflow-x: auto;
        white-space: pre-wrap;
        word-break: break-word;
        border: 1px solid rgba(51, 65, 85, 0.6);
      }}
      .tool-title {{ margin: 0 0 0.35rem; }}
      .tool-tag {{
        padding: 0.15rem 0.5rem;
        background: rgba(8, 145, 178, 0.25);
        color: #a5f3fc;
        border-radius: 6px;
        font-size: 0.88rem;
      }}
      .fold {{
        margin-top: 0.45rem;
        color: #94a3b8;
      }}
      .fold summary {{
        cursor: pointer;
        font-size: 0.78rem;
        color: #a5b4fc;
      }}
      .card footer {{
        margin-top: 0.55rem;
        font-size: 0.72rem;
        color: #475569;
        border-top: 1px dashed rgba(71, 85, 105, 0.5);
        padding-top: 0.45rem;
      }}
    </style>
    <div class="eve-trace">
      <h2>Agent execution trace</h2>
      <p class="q"><strong>Query</strong> — {_escape(query)}</p>
      <div class="gantt-cap">
        <span>Wall timeline (by step latency)</span>
        <span>{_escape(f"{total_wall:.1f}s")} total</span>
      </div>
      <div class="gantt">{"".join(gantt_rows)}</div>
      <div class="timeline">{"".join(cards)}</div>
    </div>
    """
    display(HTML(html_doc))


# Example: visualize the last agentic response
visualize_agent_trace(answer)
````

# Conclusion — What We Built and Where to Go Next

## What We Learned

We moved from **concepts** (domain LLMs, tools, ReAct agents) to **working integrations**: turning real geospatial services into callable units, exposing them through **MCP**, registering them in **EVE's Tool Registry**, and driving them from **EVE's agentic API** with inspectable **traces**.

Along the way, this notebook covered:

1. **From services to tools**: Why small, composable connectors beat monolithic pipelines; how [Nominatim](https://nominatim.openstreetmap.org/) turns place names into bounding boxes that downstream EO services expect; design rules (clear names, simple inputs, structured JSON output).
2. **MCP**: The [Model Context Protocol](https://modelcontextprotocol.io/) as a standard way for models to discover and invoke tools; [FastMCP](https://gofastmcp.com/) to generate schemas and run a server over **HTTP** (or stdio for desktop clients).
3. **EVE Tool Registry**: Community and ESA Φ-lab **MCP servers** (for example `effis`, `eve_retrieval`), authentication, listing servers, and calling a deployed tool from the platform.
4. **Agentic EVE**: The [ReAct](https://arxiv.org/abs/2210.03629) loop (reason → act with a tool → observe) implemented in practice via EVE's **LangChain / LangGraph** stack; experimental endpoints `generate-agentic` and `stream-generate-agentic` with `public_mcp_servers` to scope which tools the model may use.
5. **Tracing**: Reading the returned **trace** (agent steps vs tool steps, latencies, arguments) so you can debug ambiguous queries (for example disambiguating "Attica") and verify tool use.

The recurring idea: **capabilities** are fixed in the tool layer; **orchestration** is chosen at runtime by the model, conditioned on your question and the tool descriptions.
